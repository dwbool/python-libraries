# Deleting Class column and creating X dataset

```python
X = card.drop(columns='Class')
```

```python
X
```

# Creating y dataset

```python
y = card['Class']   
```

```python
y
```

# Splitting into training and test datasets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_train
```

```python
y_train_arr = y_train  
```

## Setting parameters

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

## Creating RandomForestClassifier model 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier 

model_grid = GridSearchCV(  estimator=RandomForestClassifier(random_state=100),
param_grid=parameters,
scoring='roc_auc',
cv=3
)
```

# Training model

```python
model_grid.fit(X_train, y_train_arr  )

```

## The best model parameters

```python
model_grid.best_params_
```

```python
model_grid
```

## Predicting by the best model

```python
proba_all = model_grid.predict_proba(X_test)
proba_all
```

```python
y_pred_proba = proba_all[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score

# https://towardsdatascience.com/how-to-calculate-roc-auc-score-for-regression-models-c0be4fdf76bb

# Calculating roc_uuc_score

roc_auc_score( y_test , y_pred_proba)
```

## Calculating AUC

```python
import sklearn.metrics as metrics
# calculate the fpr and tpr for all thresholds of the classification
#probs = model.predict_proba(X_test)
#preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred_proba)
roc_auc = metrics.auc(fpr, tpr)


print ("AUC = %.16f" % ( roc_auc ) )
```

```python
fpr
```

## Plotting data 

```python

# Plotting data 

import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
 
```
