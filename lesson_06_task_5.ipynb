## Calculating correlation matrix

```python
X_corr  = X.corr()
X_corr
```

## Finding correlation features with the value above 0.5

```python
high_corr = []

# Assuming our matrix is square shaped, can use columns length as row number

for i in range(0,len(X_corr.columns)-0 ):
    #print( str(list(X_corr['target'])[i] )+ ' ' + list( X_corr.columns)[i])
    
    corr_i = list(X_corr['target'])[i]
    
    #list( X_corr.columns)[i] 
    if corr_i > 0.5 :
        s = list( X_corr.columns)[i]
        print('found '+ str(i)+': '+ str(list(X_corr['target'])[i] )+ ' = ' + list( X_corr.columns)[i])
        if s.strip().lower() != 'target':
            high_corr.append( s )
            
        
         
high_corr        
```

```python
# List of high correlation features

high_corr
```

```python
df2
```

## Removing target column and adding High Correlation squared columns

```python

X.drop(columns=['target'])


for corr_name in high_corr:
    lst = []
    print('adding '+corr_name+'^2')
    
    for i in range(0,len(X)):
        
        val = X[corr_name][i]
        #print( val , val**2 )
        
        lst.append(val**2)
        
    X.insert( len(X.columns) , corr_name+'_2' , lst, True)
    
    
# Resulting dataset:

X
```
